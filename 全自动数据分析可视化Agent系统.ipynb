&emsp;&emsp;ä¸ºäº†å®ç°ç”Ÿæˆ`SQL`å¹¶å®é™…æŸ¥è¯¢æ•°æ®çš„çœŸå®ä¸šåŠ¡æµç¨‹ï¼Œæˆ‘ä»¬å…ˆåˆ›å»ºä¸€ä¸ªç¤ºä¾‹æ•°æ®åº“**company.db**ç”¨æ¥å­˜å‚¨ç”µå•†è®¢å•æ•°æ®åº“ï¼Œå…¶ä¸­åŒ…å«ï¼š

- `employees` è¡¨ï¼šå‘˜å·¥ä¿¡æ¯
- `products` è¡¨ï¼šäº§å“ä¿¡æ¯
- `orders` è¡¨ï¼šè®¢å•è®°å½•

&emsp;&emsp;å¹¶æ’å…¥ä¸€äº›ç¤ºä¾‹æ•°æ®ç”¨äºæµ‹è¯•æŸ¥è¯¢ã€‚

````python
# åˆ›å»º SQLite æ•°æ®åº“
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# åˆ›å»ºå‘˜å·¥è¡¨
cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    department TEXT NOT NULL,
    salary REAL NOT NULL,
    hire_date TEXT
)
""")

# åˆ›å»ºäº§å“è¡¨
cursor.execute("""
CREATE TABLE IF NOT EXISTS products (
    id INTEGER PRIMARY KEY,
    product_name TEXT NOT NULL,
    category TEXT NOT NULL,
    price REAL NOT NULL,
    stock INTEGER DEFAULT 0
)
""")

# åˆ›å»ºè®¢å•è¡¨
cursor.execute("""
CREATE TABLE IF NOT EXISTS orders (
    id INTEGER PRIMARY KEY,
    employee_id INTEGER,
    product_id INTEGER,
    quantity INTEGER,
    order_date TEXT,
    FOREIGN KEY (employee_id) REFERENCES employees (id),
    FOREIGN KEY (product_id) REFERENCES products (id)
)
""")

# æ’å…¥å‘˜å·¥æ•°æ®
employees_data = [
    (1, "å¼ ä¸‰", "æŠ€æœ¯éƒ¨", 15000, "2023-01-15"),
    (2, "æå››", "é”€å”®éƒ¨", 12000, "2023-02-20"),
    (3, "ç‹äº”", "æŠ€æœ¯éƒ¨", 18000, "2022-11-10"),
    (4, "èµµå…­", "äººåŠ›èµ„æºéƒ¨", 10000, "2023-03-01"),
    (5, "é’±ä¸ƒ", "é”€å”®éƒ¨", 16000, "2022-12-05"),
]

# æ’å…¥äº§å“æ•°æ®
products_data = [
    (1, "ç¬”è®°æœ¬ç”µè„‘", "ç”µå­äº§å“", 5999.0, 50),
    (2, "æœºæ¢°é”®ç›˜", "ç”µå­äº§å“", 499.0, 100),
    (3, "åŠå…¬æ¤…", "åŠå…¬ç”¨å“", 899.0, 30),
    (4, "æ˜¾ç¤ºå™¨", "ç”µå­äº§å“", 1999.0, 40),
]

# æ’å…¥è®¢å•æ•°æ®
orders_data = [
    (1, 1, 1, 2, "2024-01-15"),
    (2, 2, 2, 5, "2024-01-16"),
    (3, 3, 1, 1, "2024-01-17"),
    (4, 5, 3, 3, "2024-01-18"),
    (5, 2, 4, 2, "2024-01-19"),
]

cursor.executemany(
    "INSERT OR REPLACE INTO employees VALUES (?, ?, ?, ?, ?)", employees_data
)
cursor.executemany(
    "INSERT OR REPLACE INTO products VALUES (?, ?, ?, ?, ?)", products_data
)
cursor.executemany(
    "INSERT OR REPLACE INTO orders VALUES (?, ?, ?, ?, ?)", orders_data
)

conn.commit()
conn.close()

print("æ•°æ®åº“åˆ›å»ºæˆåŠŸï¼")
print("å·²åˆ›å»º 3 å¼ è¡¨ï¼šemployees, products, orders")
````

&emsp;&emsp;æ¥ä¸‹æ¥åˆå§‹åŒ– SQL Databaseã€‚é€šè¿‡ä½¿ç”¨ LangChain çš„ `SQLDatabase` è¿æ¥æ•°æ®åº“ï¼Œè¯¥å¯¹è±¡å¯ä»¥æä¾›:

````python
# è¿æ¥æ•°æ®åº“
db = SQLDatabase.from_uri("sqlite:///company.db")

print("æ•°æ®åº“ä¿¡æ¯ï¼š")
print(f"å¯ç”¨è¡¨: {db.get_usable_table_names()}")
print(f"\nç¤ºä¾‹æŸ¥è¯¢ç»“æœ:")
print(db.run("SELECT * FROM employees LIMIT 3;"))
````

&emsp;&emsp;åˆå§‹åŒ– DeepSeek æ¨¡å‹ï¼š

````python
# åˆå§‹åŒ– DeepSeek æ¨¡å‹
model = ChatDeepSeek(
    model="deepseek-chat",  # DeepSeek-V3
    temperature=0,          # ç¡®å®šæ€§è¾“å‡º
    max_retries=2,
)

print(" DeepSeek æ¨¡å‹åˆå§‹åŒ–æˆåŠŸï¼")
````

&emsp;&emsp;åˆ›å»º SQL å·¥å…·åŒ…ã€‚è¿™é‡Œä½¿ç”¨`SQLDatabaseToolkit` åˆ›å»º SQL Agent æ‰€éœ€çš„å·¥å…·é›†ã€‚è¯¥å·¥å…·é›†åŒ…å«ï¼š

1. `sql_db_query` - æ‰§è¡Œ SQL æŸ¥è¯¢
2. `sql_db_schema` - è·å–è¡¨ç»“æ„
3. `sql_db_list_tables` - åˆ—å‡ºæ‰€æœ‰è¡¨
4. `sql_db_query_checker` - SQL è¯­æ³•æ£€æŸ¥

````python
# åˆ›å»º SQL å·¥å…·åŒ…
toolkit = SQLDatabaseToolkit(db=db, llm=model)

# è·å–æ‰€æœ‰å·¥å…·
tools = toolkit.get_tools()

print("SQL å·¥å…·åŒ…åˆ›å»ºæˆåŠŸï¼")
print(f"\nå¯ç”¨å·¥å…· ({len(tools)} ä¸ª)ï¼š")
for tool in tools:
    print(f"  â€¢ {tool.name}: {tool.description}...")
````

&emsp;&emsp;æœ€åï¼Œä½¿ç”¨ LangChain 1.0 çš„ `create_agent` åˆ›å»ºæ™ºèƒ½ SQL Agentã€‚

````python
# åˆ›å»º SQL Agent
agent = create_agent(
    model=model,
    tools=tools,
    system_prompt="""ä½ æ˜¯ä¸€ä¸ªä¸“ä¸šçš„ SQL æ•°æ®åˆ†æå¸ˆã€‚

å·¥ä½œæµç¨‹ï¼š
1. å…ˆä½¿ç”¨ sql_db_list_tables æŸ¥çœ‹æ‰€æœ‰è¡¨
2. ä½¿ç”¨ sql_db_schema è·å–ç›¸å…³è¡¨çš„ç»“æ„
3. ç”Ÿæˆ SQL æŸ¥è¯¢å‰ï¼Œä½¿ç”¨ sql_db_query_checker æ£€æŸ¥è¯­æ³•
4. ä½¿ç”¨ sql_db_query æ‰§è¡ŒæŸ¥è¯¢
5. ç”¨ä¸­æ–‡æ€»ç»“æŸ¥è¯¢ç»“æœ

æ³¨æ„äº‹é¡¹ï¼š
- åªä½¿ç”¨æ•°æ®åº“ä¸­å®é™…å­˜åœ¨çš„è¡¨å’Œå­—æ®µ
- æŸ¥è¯¢ç»“æœé™åˆ¶åœ¨ 10 æ¡ä»¥å†…
- å¦‚æœæŸ¥è¯¢å‡ºé”™ï¼Œåˆ†æé”™è¯¯å¹¶é‡æ–°ç”Ÿæˆ SQL
""",
)

print("SQL Agent åˆ›å»ºæˆåŠŸï¼")
````

&emsp;&emsp;ä»…éœ€è¦å¦‚ä¸Šæ“ä½œï¼Œç°åœ¨ä¾¿å¯ä»¥å¼€å§‹ç”¨è‡ªç„¶è¯­è¨€æé—®äº†ï¼Agent ä¼šè‡ªåŠ¨å‘ç°ç›¸å…³è¡¨ç»“æ„ï¼Œç”Ÿæˆå¹¶éªŒè¯ SQLï¼Œæ‰§è¡ŒæŸ¥è¯¢ï¼Œè¿”å›ç»“æœã€‚

````python
# ç¤ºä¾‹ 1ï¼šç®€å•æŸ¥è¯¢
question1 = "æœ‰å¤šå°‘åå‘˜å·¥ï¼Ÿ"

print(f"é—®é¢˜: {question1}")
response1 = agent.invoke({"messages": [{"role": "user", "content": question1}]})

# æ‰“å°æœ€ç»ˆå›ç­”
final_message = response1["messages"][-1]
print(f"\nå›ç­”: {final_message.content}")
print("\n" + "="*60)
````
